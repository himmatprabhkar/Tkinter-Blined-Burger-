```python
from tkinter import*
import random
import time
from datetime import datetime
from tkinter import messagebox

window=Tk()
window.configure(background='white')
window.geometry('1350x650+0+0')
window.title("Billing System")

#=================================Design Part================================================================
Tops = Frame(window, width=1350, height=100, bd=8, relief="raise" ,background="blue")
Tops.pack(side=TOP)

f1 = Frame(window, width=900, height=650, bd=8, relief="raise",background="blue")
f1.pack(side=LEFT)
f2 = Frame(window, width=450, height=650, bd=8, relief="raise" ,background="blue")
f2.pack(side=LEFT)

f1a = Frame(f1, width=900, height=330, bd=8, relief="raise")
f1a.pack(side=TOP)
f2a = Frame(f1, width=900, height=320, bd=8, relief="raise")
f2a.pack(side=BOTTOM)

f1aa = Frame(f1a, width=400, height=430, bd=8, relief="raise")
f1aa.pack(side=LEFT)
f1ab = Frame(f1a, width=400, height=430, bd=8, relief="raise")
f1ab.pack(side=RIGHT)

f2aa = Frame(f2a, width=450, height=320, bd=8, relief="raise")
f2aa.pack(side=LEFT)
f2ab = Frame(f2a, width=450, height=320, bd=8, relief="raise", background="yellow")
f2ab.pack(side=LEFT)
#=============================================================================================================

lblinfo = Label(Tops, font=('arial',30,'bold'),text="         Welcome To Burger King       ",bd=10, anchor="w",
                bg="orange", fg="red")
lblinfo.grid(row=0,column=0)

#=========================Variables Convert int to str===================================================================
text_input=StringVar()
operator=""
PaymentRef=StringVar()
SalesReference=StringVar()
Carpets=StringVar()
Fabric=StringVar()
Blinds=StringVar()
HomeDelivery=StringVar()
CostofCarpets=StringVar()
CostofFabric=StringVar()
CostofBlinds=StringVar()
CostofDelivery=StringVar()
PaidTax=StringVar()
SubTotal=StringVar()
TotalCost=StringVar()
DateofOrder=IntVar()

#==============================================For the Rest Function===================================================
Carpets.set(0)
Fabric.set(0)
Blinds.set(0)
HomeDelivery.set(0)
DateofOrder.set(time.strftime("%d/%m/%y"))
#==============================================Function For The Calculator and Buttons==================================
def iExit():
    qExit = messagebox.askyesno("Billing System", "Do You Want To Exit The System")
    if qExit > 0:
        window.destroy()
        return

def PayReference():
    x = random.randint(10034, 699812)
    randomRef = str(x)
    PaymentRef.set("BILL" +  randomRef)
def btnClick(numbers):
    global operator
    operator = operator + str(numbers)
    text_input.set(operator)
def btnClearDisplay():
    global operator
    operator=""
    text_input.set("")
def btnEqualsInput():
    global operator
    sumup =str(eval(operator))
    text_input.set(sumup)
    operator=""
    
def Reset():
    PaymentRef.set("")
    SalesReference.set("")
    Carpets.set(0)
    Fabric.set(0)
    Blinds.set(0)
    HomeDelivery.set(0)
    CostofCarpets.set("")
    CostofFabric.set("")
    CostofBlinds.set("")
    CostofDelivery.set("")
    PaidTax.set("")
    SubTotal.set("")
    TotalCost.set("")
    
def CostofOrder():
    CarpetsPrice = float(Carpets.get())
    BlindsPrice = float(Blinds.get())
    FabricPrice = float(Fabric.get())
    DeliveryCost= float(HomeDelivery.get())
    
    CarpetsCost = "RS" , str("%.2f" %(( CarpetsPrice * 15.49)))
    CostofCarpets.set(CarpetsCost)
    
    BlindsCost = "RS" , str("%.2f" %(( BlindsPrice * 7.49)))
    CostofBlinds.set(BlindsCost)
    
    FabricCost = "RS" , str("%.2f" %(( FabricPrice * 5.50)))
    CostofFabric.set(FabricCost)
    
    Delivery = "RS" , str("%.2f" %(( DeliveryCost * 4.50)))
    CostofDelivery.set(Delivery)

    ToC = "RS" , str("%.2f" %(( CarpetsPrice * 15.49) + ( BlindsPrice * 7.49) + ( FabricPrice * 5.50) + ( DeliveryCost * 4.50)))
    SubTotal.set(ToC)
    Tax = "RS" , str("%.2f" %(( CarpetsPrice * 15.49) + ( BlindsPrice * 7.49) + ( FabricPrice * 5.50) + ( DeliveryCost * 4.50)))
    PaidTax.set(Tax)
    TaxPay = ((( CarpetsPrice * 15.49) + ( BlindsPrice * 7.49) + ( FabricPrice * 5.50) + ( DeliveryCost * 4.50)) * 0.2)
    Cost =  ((( CarpetsPrice * 15.49) + ( BlindsPrice * 7.49) + ( FabricPrice * 5.50) + ( DeliveryCost * 4.50)))
    CostofItems = "RS" , str("%.2f" % ( TaxPay + Cost))
    TotalCost.set(CostofItems)
                             
    x = random.randint(10034, 699812)
    randomRef = str(x)
    PaymentRef.set("BILL" +  randomRef)

#=============================================================================================================
textdisplay = Entry(f2, font=('arial',20,'bold'),textvariable=text_input, bd=40 , insertwidth=6, justify="right")
textdisplay.grid(columnspan=4)

#=============================================================================================================
btn7 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="7",command=lambda:btnClick(7))
btn7.grid(row=1,column=0)
btn8 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="8",command=lambda:btnClick(8))
btn8.grid(row=1,column=1)
btn9 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="9",command=lambda:btnClick(9))
btn9.grid(row=1,column=2)
btnPlus = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="+",command=lambda:btnClick("+"))
btnPlus.grid(row=1,column=3)
#=============================================================================================================
btn4 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="4",command=lambda:btnClick(4))
btn4.grid(row=3,column=0)
btn5 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="5",command=lambda:btnClick(5))
btn5.grid(row=3,column=1)
btn6 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="6",command=lambda:btnClick(6))
btn6.grid(row=3,column=2)
btnSub = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="-",command=lambda:btnClick("-"))
btnSub.grid(row=3,column=3)
#=============================================================================================================
btn1 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="1",command=lambda:btnClick(1))
btn1.grid(row=4,column=0)
btn2 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="2",command=lambda:btnClick(2))
btn2.grid(row=4,column=1)
btn3 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="3",command=lambda:btnClick(3))
btn3.grid(row=4,column=2)
btnMul = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="*",command=lambda:btnClick("*"))
btnMul.grid(row=4,column=3)
#=============================================================================================================
btn7 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="0",command=lambda:btnClick(0))
btn7.grid(row=5,column=0)
btn8 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="C",command=btnClearDisplay)
btn8.grid(row=5,column=1)
btn9 = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="=",command=btnEqualsInput)
btn9.grid(row=5,column=2)
btnDiv = Button(f2,padx=16, pady=16,bd=8, fg="black" ,font=('arial',20,'bold'),text="/",command=lambda:btnClick("/"))
btnDiv.grid(row=5,column=3)

#============================================Order Information flaa===============================================

lblRef=Label(f1aa,font=("arial",16,"bold"),text="Sales Reference", bd=16, justify="left")
lblRef.grid(row=0,column=0)
textRef=Entry(f1aa,font=("arial",16,"bold"),textvariable=PaymentRef, bd=10,insertwidth=2)
textRef.grid(row=0,column=1)

lblRef=Label(f1aa,font=("arial",16,"bold"),text="Carpets", bd=16, justify="left")
lblRef.grid(row=1,column=0)
textRef=Entry(f1aa,font=("arial",16,"bold"),textvariable=Carpets, bd=10,insertwidth=2)
textRef.grid(row=1,column=1)
 

lblRef=Label(f1aa,font=("arial",16,"bold"),text="Fabric", bd=16, justify="left")
lblRef.grid(row=2,column=0)
textRef=Entry(f1aa,font=("arial",16,"bold"),textvariable=Fabric, bd=10,insertwidth=2)
textRef.grid(row=2,column=1)

lblRef=Label(f1aa,font=("arial",16,"bold"),text="Blinds", bd=16, justify="left")
lblRef.grid(row=3,column=0)
textRef=Entry(f1aa,font=("arial",16,"bold"),textvariable=Blinds, bd=10,insertwidth=2)
textRef.grid(row=3,column=1)

lblRef=Label(f1aa,font=("arial",16,"bold"),text="Home Delivery", bd=16, justify="left")
lblRef.grid(row=4,column=0)
textRef=Entry(f1aa,font=("arial",16,"bold"),textvariable=HomeDelivery, bd=10,insertwidth=2)
textRef.grid(row=4,column=1)

#============================================Order Information flab===============================================

lblRef=Label(f1ab,font=("arial",16,"bold"),text="Order Date", bd=16, justify="left")
lblRef.grid(row=0,column=0)
textRef=Entry(f1ab,font=("arial",16,"bold"),textvariable=DateofOrder, bd=10,insertwidth=2)
textRef.grid(row=0,column=1)
lblRef=Label(f1ab,font=("arial",16,"bold"),text="Cost of Carpets", bd=16, justify="left")
lblRef.grid(row=1,column=0)
textRef=Entry(f1ab,font=("arial",16,"bold"),textvariable=CostofCarpets, bd=10,insertwidth=2)
textRef.grid(row=1,column=1)


lblRef=Label(f1ab,font=("arial",16,"bold"),text="Cost of Fabric", bd=16, justify="left")
lblRef.grid(row=2,column=0)
textRef=Entry(f1ab,font=("arial",16,"bold"),textvariable=CostofFabric,bd=10,insertwidth=2)
textRef.grid(row=2,column=1)

lblRef=Label(f1ab,font=("arial",16,"bold"),text="Cost of Blinds", bd=16, justify="left")
lblRef.grid(row=3,column=0)
textRef=Entry(f1ab,font=("arial",16,"bold"),textvariable=CostofBlinds, bd=10,insertwidth=2)
textRef.grid(row=3,column=1)


lblRef=Label(f1ab,font=("arial",16,"bold"),text="Cost of Delivery", bd=16, justify="left")
lblRef.grid(row=4,column=0)
textRef=Entry(f1ab,font=("arial",16,"bold"),textvariable=CostofDelivery, bd=10,insertwidth=2)
textRef.grid(row=4,column=1)

#============================================Order Information f2aa=============================================

lblRef=Label(f2aa,font=("arial",16,"bold"),text="Paid Text", bd=16, justify="left")
lblRef.grid(row=0,column=0)
textRef=Entry(f2aa,font=("arial",16,"bold"),textvariable=PaidTax, bd=10,insertwidth=2)
textRef.grid(row=0,column=1)

lblRef=Label(f2aa,font=("arial",16,"bold"),text="Sub Total", bd=16, justify="left")
lblRef.grid(row=1,column=0)
textRef=Entry(f2aa,font=("arial",16,"bold"),textvariable=SubTotal, bd=10,insertwidth=2)
textRef.grid(row=1,column=1)


lblRef=Label(f2aa,font=("arial",16,"bold"),text="Total Cost", bd=16, justify="left")
lblRef.grid(row=2,column=0)
textRef=Entry(f2aa,font=("arial",16,"bold"),textvariable=TotalCost, bd=10,insertwidth=2)
textRef.grid(row=2,column=1)

# #============================================Order Information Buttons=============================================

btnTotal = Button(f2ab,padx=16, pady=16,bd=7, fg="black" ,font=('arial',10,'bold'),width=20,text="Total Cost",command=CostofOrder)
btnTotal.grid(row=0,column=0)
btnRefer = Button(f2ab,padx=16, pady=16,bd=7, fg="black" ,font=('arial',10,'bold'),width=20,text="Seles Reference",command=PayReference)
btnRefer.grid(row=0,column=1)
btnRest = Button(f2ab,padx=16, pady=16,bd=7, fg="black" ,font=('arial',10,'bold'),width=20,text="Rest", command=Reset)
btnRest.grid(row=1,column=0)
btnExit = Button(f2ab,padx=16, pady=16,bd=7, fg="black" ,font=('arial',10,'bold'),width=20,text="Exit", command=iExit)
btnExit.grid(row=1,column=1)
          
window.mainloop()
```
